# Import Libraries

```python
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt
%matplotlib inline
```

# Load Dataset

```python
df = pd.read_csv("../data/unlalbeled/data_invoice.csv")
display(df.head(), df.tail())
```

```python
df = df.dropna()
display(df.head(), df.tail())
```

# Data Exploration

```python
def plot_dataset_distribution(X, title):
    plt.scatter(X[:,0], X[:,1])
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(df.to_numpy())
```

```python
plot_dataset_distribution(df.to_numpy(), "UX Custumer Segmentation")
```

```python
tsne = TSNE(n_components=2,
            init='pca',
            random_state=42)
X_tsne = tsne.fit_transform(X_std)
```

```python
plot_dataset_distribution(X_tsne, "UX Custumer Segmentation")
```
