# Linear Algebra Assignment

## Question 1: Compute the distance between:




```python
import numpy as np
from scipy.linalg import svd
```

### a) $x^{T}y$

```python
x = np.array([1, 2, 3]).T
y = np.array([-1, -1, 0]).T
print(np.matmul(x.T,y))
```

### b) $x^{T}Ay$

```python
A = np.array([[2, 1, 0],
              [1, 3, -1],
              [0, -1, 2]])
print(A)
```

```python
print(np.matmul(np.matmul(x.T,A), y))
```

## Question 2: Compute the angle between



```python
x = np.array([1, 2]).T
y = np.array([-1, -1]).T
B = np.array([[2, 1],
              [1, 3]])
```

### a) $x^{T}y$

```python
print(np.matmul(x.T,y))
```

### b) $x^{T}By$

```python
print(np.matmul(np.matmul(x.T, B), y))
```

## Question 3: Compute the SVD (Singular Value Decomposition) of the following matrix and present the interpretation of each matrix:



```python
A = np.array([[3, 2, 2],
              [2, 3, -2]])
# A = np.array([[5, 4, 1],
#               [5, 5, 0],
#               [0, 0, 5],
#               [1, 0, 4]])
A.shape
```

```python
A
```

```python
U, s, VT = svd(A,)
sigma = np.zeros((A.shape[0], A.shape[1]))
for i in range(min(A.shape[0], A.shape[1])):
    sigma[i, i] = s[i]
U.shape, sigma.shape, VT.shape
```

```python
U # U also called left-singular value, its related to rows, and tells us that the samples has the same importance/correlation.
```

```python
sigma # The sigma is diagonal matrix, and tells us that the first and second column of the matrix A has more importance
```

```python
VT # VT also called right-singular tell us that the first and second row have large absolute values.
# (e.g. the first row)
```
