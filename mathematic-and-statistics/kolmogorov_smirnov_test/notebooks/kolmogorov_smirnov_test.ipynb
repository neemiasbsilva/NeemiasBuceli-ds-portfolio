# Setup & Import Libraries

```python
import numpy as np
import pandas as pd
from scipy import stats

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Generate random samples

## Normal Distribution

```python
def generate_random_samples(norm=True, parameters=[{"mean": 0, "std": 1, "size": 500},
                                                   {"mean": 0.1, "std": 1, "size": 500},
                                                   {"mean": 3, "std": 1, "size": 500}]):
    norm_samples = []
    for map in parameters:
        if (norm):
            norm_samples.append(np.random.normal(loc=map["mean"], scale=map["std"], 
                                                 size=map["size"]))
        else:
            norm_samples.append(np.random.f(dfnum=map["mean"], dfden=map["std"], size=map["size"]))
    return norm_samples
```

```python
norm_samples = generate_random_samples()
```

## F Distribution

```python
f_samples = generate_random_samples(norm=False, parameters=[{"mean": 5, "std": 15, "size": 500}])
```

## Visualize the distributions

```python
def plot_distributions(distribution, color):
    sns.histplot(distribution, bins=10, kde=True, color=color)

plt.figure(figsize=(10, 5))
colors = ["red", "blue", "green", "purple"]
legends = []
idx = 0
for distribution in norm_samples:
    plot_distributions(distribution, colors[idx])
    legends.append(f"normal_{idx+1}_dist")
    idx += 1

for i, distribution in enumerate(f_samples):
    plot_distributions(distribution, colors[idx])
    idx += 1
    legends.append(f"f_{i+1}_dist")
plt.legend(legends)
plt.plot()
```
